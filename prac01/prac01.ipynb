### Основные формулы
Если элементарные исходы равновероятны

$$P(A) = \frac{m}{n} = \frac{\textrm{число благоприятных исходов}}{\textrm{число всех исходов}}$$

Если элементарные исходы неравновероятны

$$P(A) = \frac{measure(A)}{measure(\Omega)}$$

*measure* - некая мера, $\Omega$ - множество всех элементарных событий


Число способов выбрать $k$ объектов из $n$ различных объектов.
$$C_{n}^{k} = \frac{n!}{k!(n-k)!}$$

Число перестановок
$$P_n = n!$$
