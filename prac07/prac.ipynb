## Теория <a class="anchor" id="chapter1"></a>

**Математическое ожидание**
$$\mathbb{E}[\xi] = \sum_{i}x_iP(\xi=x_i)$$

P.S. Матожидание **существует**, если оно корректно определено и конечно. 

Самое важное свойство всея тервера: для **произвольных** (зависимых и нет) сл.в. $\xi$ и $\mu$ верно, что

$$\mathbb{E}[a\xi + b\mu] = a\mathbb{E}[\xi] + b\mathbb{E}[\mu]$$

	числа остановок лифта.

## ДЗ <a class="anchor" id="chapter3"></a>

**1.** Из множества 10-буквенных слов над русским алфавитом случайным образом выбирается одно $w$ (все слова равновероятны). Найдите ожидание количества подслов “ТЕРВЕР” у слова $w$. (Подсловом считается подпоследовательность, элементы которой в исходном слове идут подряд.)

**2.** Ваша хеш-функция переводит объект число от 1 до 10000 с равной вероятностью. У нас имеется 1000 объектов. 
* (0.5) Какова вероятность того, что произойдет коллизия? 
* (0.5) Найдите матожидание кол-ва коллизий? 

**3.** У нас в хороводе стоит n куриц. В какой-то момент все курицы клюют одного из своих соседей слева или справа. Найти матожидние кол-ва куриц, которых не клюнули.

**4.** В Швейцарском Фейсбуке очень любят сотрудников, поэтому решили, устраивать всем выходной, когда у кого-нибудь из людей день рождение. Однако в этом случае они не могут набирать много людей, так как с большой вероятностью работа будет не выполнена. Работник рождается в любой из дней в году (365 дней) с равной веротяностью. Найдите наибольшее матожидание величины (кол-во сотрудников * кол-во рабочих дней).

P.S. Надо найти при каком количестве сотрудников матожидание величины будет максимальным и вывести это значение.


**5.** На день рождения Дианы мы принесли торт с N свечами. Когда мы пытаемся задуть свечи, мы точно задуваем одну свечу, а количество оставшихся гореть свечей распределено равновероятно. (от 0 до N-1). При следующем задуве ситуация аналогичная. Вопрос: найти матожидание числа задуваний, которое должна сделать Диана.

## Гробы <a class="anchor" id="chapter4"></a>

**1. (2 балла)** В социальной сети зарегестрировано конечное число пользователей.

**2. (1 балл)** Дискретная случайная величина $\xi$ принимает $k$
	положительных значений $x_1,\ldots,x_k$ с вероятностями,
	равными соответственно $p_1,\ldots,p_k$. Предполагая, что
	возможные значения записаны в возрастающем порядке,
	доказать, что
	$$\lim_{n\rightarrow\infty}\frac{E\xi^{n+1}}{E\xi^n}=x_k. $$
