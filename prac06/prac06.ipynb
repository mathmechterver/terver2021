# 11. Марковские цепи. Марковский случайный процесс.

![title](data/markov.png)

Пусть $X$ - конечное или счетное множество (фазовое пространство $\in \mathcal{N}$).

Система принимает состояния из $X$, $\xi_n$ - ее состояние в момент времени $n$ ($\xi_n: \Omega \to X$)

* Система не зависит от предыдущих состояний $ \forall n \geq 1$ и $x_1, \ldots, x_{n+1} \in X$

$$P(\xi_{n+1}=x_{n+1}|\xi_0=x_0, \xi_1=x_1, \ldots, \xi_n=x_n) = P(\xi_{n+1}=x_{n+1}|\xi_n=x_n)$$


Вероятность перехода $P_n(x,y)$

1. $\forall n \in \mathbb{N}$ $\forall x \in X$ $\sum_{y \in X} P_n(x,y) = 1$
2. $\forall n \in \mathbb{N}$ $\forall x,y \in X$ $P_n(x,y) \geq 0$
3. $P_n(x,y) = P(\xi_n = y|\xi_{n-1} = x)$, если $P(\xi_{n-1}=x)>0$ 

### Теорема. (Можно доказать по индукции [ref](https://stepik.org/lesson/254734/step/9?unit=234942))

Марковскую цепь также можно задать через след выражение и оно эквивалентно определению выше
$P(\xi_0 = x_0, \xi_1 = x_1,\ldots, \xi_n = x_n) = P(\xi_0=x_0)P_1(x_1, x_2)\ldots P_n(x_{n-1}, x_n)$

### Однородные цепи маркова

Марковская цепь является однородной, если вероятность перехода из состояния на шаге $n$ в состояние на шаге $m$ зависит только от $n-m$

$P_{n, n+1}(x,y) = P(x,y) = T_{x,y}$ - элемент матрицы перехода


### Стационарное распределение

$\pi^*$ называется стационарным, если для любого начального распределения $\pi$:
$$\lim_{n\to \infty} \pi * T^n = \pi^*$$

### Возвратные состаяния

$f_a(n) = P(\xi_n = a, \xi_{n-1} \neq a, \ldots, \xi_{1} \neq a|\xi_{0} = a)$ - вероятность возврата

* Если $\sum_{n=1}^{\infty}f_a(n)=1$, то $a$ - возвратное
* Если $\sum_{n=1}^{\infty}f_a(n)<1$, то $a$ - невозвратное
