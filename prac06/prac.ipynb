## Теория <a class="anchor" id="chapter1"></a>

Отображение $\xi: \Omega \to \mathbb{R}$ называется **Дискретной случайной величиной**, если оно принимает не более, чем счетное число значений $x_1, x_2, \ldots x_n, \ldots$, и выполнено утверждение $H_i = \{\omega \in \Omega | \xi(\omega) = x_i\} \in \mathcal{F}$

Примеры: 
* $\xi_1$ - игральный кубик $\xi_1(\omega_2) = \frac{1}{6}$
* $\xi_2$ - остаток при делении на 2, от значения на грани кубика
* 
$\xi_3 =\left\{
	\begin{array}{cc}
	12, & \xi_1 = 1,2,3\\
	24, & \xi_1 = 4,5\\
    36, & \xi_1 = 6\\
	\end{array}\right.$

**Дискретным распределением** называют отображение из счетного подмножества множества $\mathbb{R}$ в $[0,1]$ сумма элементов образа которого равна 1.

|$x_i$ | 12    | 24     |  36  |
|--------|-------|--------|------|
|$P(\xi_3 = x_i)$ |  0.5  |  0.33  | 0.17 |


**Математическое ожидание** (наконец-то)
$$\mathbb{E}[\xi] = \sum_{i}x_iP(\xi=x_i)$$

P.S. Матожидание **существует**, если оно корректно определено и конечно. 

----

Если $\xi$ принимает только целые неотрицательные значения, то можно построить **Производящую функцию**
$$\phi_{\xi}(t) = \sum_{k}t^kP(\xi=k) = \mathbb{E}[t^{\xi}]$$

## Задачи на пару <a class="anchor" id="chapter2"></a>

**1.** В партии из 10 деталей имеется 8 стандартных. Наудачу отобраны две детали. Постройте **случайную величину**, которая описывает количество стандартных деталей среди двух отобранных. Также постройте ее распределение.


**2.** У нас есть правильный кубик и неправильная монетка, на сторонах которой написаны $0$ и $1$: $0$ выпадает с вероятностью $\frac{2}{3}$, $1$ соответственно $\frac{1}{3}$. (Очевидно, но данные сл.вел., конечно же, независимы)

Пусть $\xi$ - ДСВ, которая соответствует числу, выпавшему на грани первого кубика, а $\mu$ - на неправильной монетке.

Найти **распределения** и **матожидания** случайных величин:
* $\theta = \xi +\mu + 1$
* $\theta = -2\xi\mu$
* $\theta = \frac{\mu}{2^{\xi}}$

**3.** Мы 2 раза бросили игральный кубик. Пусть $\xi_1$ - сл.в. результата 1 броска, $\xi_2$ - сл.в. результата 2 броска.

В чем разница между случайными величинами $\xi_1 \xi_1$ и $\xi_1 \xi_2$? Найдите их матожидания.

**4.** Монета выпадает орлом с вероятностью $p > 0$. Пусть $\xi$ — число подбрасываний, необходимых для достижения $1$ выпадения орла. Найти производящую функцию для $\xi$.

**5.** Доказать, что $\phi_{\xi}(1)' = \mathbb{E}[\xi]$ и найти матожидание в предыдущей задаче.
